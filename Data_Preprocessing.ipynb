```python
import pandas as pd
df=pd.read_csv('raw_cafe.csv')
display(df)
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True) 
display(df)
```

```python
df['Area']=df['Area'].str.replace(', Surat','')
```

```python
df['Dining_rating']=df['Dining_rating'].str.replace('-','unavailable')
```

```python
df['Coffee_availibility']=df['Food_Items'].str.contains('Coffee')
display(df)
```

```python
df['Average_Cost']=df['Average_Cost'].str.replace('₹','')
df['Average_Cost']=df['Average_Cost'].str.replace('for two','')
df['Average_Cost']=df['Average_Cost'].str.replace(',','')
display(df)
```

```python
df['Average_Cost']=df['Average_Cost'].astype('str').astype('int')
df['Average_Cost'].info
df['Average_Cost']=df['Average_Cost']/2
display(df)
```

```python
df.to_csv('final_cafe.csv')
```
